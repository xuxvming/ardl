```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import statsmodels.formula.api as smf
import statsmodels.tsa.api as smt
import statsmodels.api as sm
import scipy.stats as scs
import numpy as np
from statsmodels.tsa.stattools import adfuller

df = pd.read_excel('file.xls',sheet_name = "Sheet1")
df
```

```python
print (df.columns)
```

```python
df = df.drop('地区',axis =1)
```

```python
time_series ={}
for col in df.columns:
    val = df[col][0]
    col = col.replace('年','/')
    col = col.replace('月','')
    time_series[col] = val 
```

```python
df = pd.DataFrame.from_dict(time_series,orient='index')
```

```python
df.columns = ['index']
df
```

```python
def tsplot(y, lags=None, figsize=(10, 8), style='bmh'):
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):    
        fig = plt.figure(figsize=figsize)
        layout = (3, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))
        pp_ax = plt.subplot2grid(layout, (2, 1))
        
        y.plot(ax=ts_ax)
        ts_ax.set_title('Time Series Analysis Plots')
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.5)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.5)      
        scs.probplot(y, sparams=(y.mean(), y.std()), plot=pp_ax)

        plt.tight_layout()
    return
```

```python
tsplot(df['index'])
```

```python
#数据好像还行, 还可以更优化一下
```

```python
#perform Argumented Dickey Fuller Test
```

```python
adf_test = adfuller(df['index'])
print ('ADF = ' + str(adf_test[0]))
print ('p-value = ' +str(adf_test[1]))
```

```python
#log return of first difference
```

```python
df_log = np.log(df['index']/df['index'].shift(1)).dropna(how='all')
tsplot(df_log)
```

```python
#run adf again
```

```python
adf_test = adfuller(df_log)
print ('ADF = ' + str(adf_test[0]))
print ('p-value = ' +str(adf_test[1]))
```
